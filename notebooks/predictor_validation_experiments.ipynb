```python
import pandas as pd
import matplotlib.pyplot as plt

from typing import Optional
from globals import root_path
```

```python
def fig(targets: str, out: Optional[str] = None) -> None:
    plt.figure(figsize=(12, 4))
    for i, target in enumerate(targets):
        plt.subplot(131 + i)
        category = ['cv', 'ind']
        for j, legend in enumerate(['Cross Validation', 'Independent Test']):
            df = pd.read_table(root_path / f"models/RF_REG{target}.{category[j]}.tsv")
            df = df[df.Label != 3.99]
            plt.scatter(df.Label, df.Score, s=5, label=legend)
        plt.xlim([3.5, 10.5])
        plt.ylim([3.5, 10.5])
        plt.xlabel('Real pX')
        plt.ylabel('Predicted pX')
        plt.legend(loc='upper left')
    plt.tight_layout()
    if out is None:
        plt.show()
    else:
        plt.savefig(out, dpi=300)
```

```python
fig(targets=["CHEMBL226", "CHEMBL240", "CHEMBL251"])
```
